### 2. 2013_baseball.csv는 2013년 시즌 각 타자의 성적을 포함하고 있다.평행좌표 플롯을 활용해 선수들의 성적 패턴에 어떤 경향이 있는지 알아보려 한다.포지션 별,팀별 평행좌표 플롯을 그리고 각 포지션,혹은 팀별 타자의 성적에 어떤 경향이 있는지 서술하라.

```r
bb2013=read.csv("2013_baseball.csv")
head(bb2013)
```

```r
position=bb2013$포지션
```

```r
base2_pos=bb2013[,c(2,4:11)]
```

```r
base2_pos2=aggregate(base2_pos[,2:9],by=list(포지션=base2_pos$포지션), )
```

```r
rownames(base2_pos2)=base2_pos2[,1]
```

```r
library(lattice)
```

```r
parallel(~bb2013[,4:11]|position,horizontal.axis=F,col=1,family="NanumGothic")
parallel(base2_pos2[,2:9],horizontal.axis=F,col=1,family="NanumGothic")
```

- 팀별평행좌표

```r
team=bb2013$팀
head(team)
```

```r
parallel(~bb2013[,4:11]|team,horizontal.axis = F,col=1)
```

### 3.	2013_baseball.csv에 있는 각 타자에 대한 변수를 사용하여 주성분 분석을 시행하라.적당한 주성분의 개수를 파악하고 해당 주성분들로 설명되는 분산의 비율을 구하라.행렬도를 통해 선수들의 특징이 어떻게 파악되는지 살펴보시오.

```r
library(Deducer)
```

```r
rownames(bb2013)=bb2013[,1]
rownames(bb2013)
head(bb2013)
model=prcomp(bb2013[,4:11],scale=T)
```

```r
plot(model)
```

```r
summary(model)
```

```r
biplot(model)
```
