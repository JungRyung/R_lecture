# Multiple Logistic Regression
## Disease Outbreak

```r
data = read.csv("disease.csv")
```

```r
# full model
model3 = glm(disease~. , data=data, family = binomial(logit))
summary(model3)
```

```r
# reduced model
model4 = glm(disease~age+sector, data=data, family = binomial(logit))
summary(model4)
```

```r
# 독립성 검증
anova(model3, model4, test='Chisq')
```
